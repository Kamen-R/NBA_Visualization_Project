```python
def get_city(name):
    words = name.split()
    
    if len(words) == 2:
        return words[0]
    else:
        return words[0] + " " + words[1]
```

```python
nicknames = nicknames.replace({"New Orleans/Oklahoma City Hornets": "New Orleans Hornets"})
```

```python
nicknames["city"] = nicknames["Name"].apply(get_city)
```

```python
nicknames = nicknames.drop(columns="Name")
```

```python
nicknames = nicknames.replace({"Golden State": "San Francisco", "Indiana": "Indianapolis", "Minnesota": "Minneapolis",
                   "Portland Trail": "Portland", "Utah": "Salt Lake City"})
```

```python
state_dict = {
    'Atlanta': "GA",'Brooklyn': "NY", 'Boston': "MA", 'Charlotte': "NC", 'Chicago': "IL",
    'Cleveland': "OH", 'Dallas': "TX", 'Denver': "CO", 'Detroit': "MI", 'San Francisco': "CA",
    'Houston': "TX", 'Indianapolis': "IN", 'Los Angeles': "CA", 'Memphis': "TN", 'Miami': "FL",
    'Milwaukee': "WI", 'Minneapolis': "MN", 'New Jersey': "Duplicate", 'New Orleans': "LA", 'New York': "NY",
    'Oklahoma City': "OK", 'Orlando': "FL", 'Philadelphia': "PA", 'Phoenix': "AZ",
    'Portland': "OR", 'Seattle': "WA", 'Sacramento': "CA", 'San Antonio': "TX",
    'Toronto': "Canada", 'Salt Lake City': "UT", 'Vancouver': "Duplicate", 'Washington': "DC"}
def get_state(city):
    return state_dict[city]
```

```python
nicknames["state_id"] = nicknames["city"].apply(get_state)
```
